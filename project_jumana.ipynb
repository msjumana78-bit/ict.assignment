```python
#Design a Mini Banking System that simulates basic banking operations.


MIN_BALANCE = 500
DAILY_LIMIT = 10000

accounts = {}

# Create Account
def create_account():
    name = input("Enter name: ")
    if name in accounts:
        print("Account already exists!")
        return

    pin = input("Set PIN: ")
    balance = float(input("Enter initial deposit: "))

    if balance < MIN_BALANCE:
        print("Minimum balance must be maintained!")
        return

    accounts[name] = {
        "pin": pin,
        "balance": balance,
        "withdraw_today": 0,
        "transactions": ["Account created with ₹" + str(balance)]
    }

    print("Account created successfully!")

# Deposit Money
def deposit(acc):
    amt = float(input("Enter deposit amount: "))
    if amt <= 0:
        print("Invalid amount!")
        return

    acc["balance"] += amt
    acc["transactions"].append("Deposited ₹" + str(amt))
    print("Deposit successful!")

# Withdraw Money
def withdraw(acc):
    amt = float(input("Enter withdrawal amount: "))

    if amt <= 0:
        print("Invalid amount!")
        return

    if acc["withdraw_today"] + amt > DAILY_LIMIT:
        print("Daily withdrawal limit exceeded!")
        return

    if acc["balance"] - amt < MIN_BALANCE:
        print("Minimum balance must be maintained!")
        return

    acc["balance"] -= amt
    acc["withdraw_today"] += amt
    acc["transactions"].append("Withdrawn ₹" + str(amt))
    print("Withdrawal successful!")

# Transfer Money
def transfer(acc):
    receiver = input("Enter receiver name: ")
    if receiver not in accounts:
        print("Receiver not found!")
        return

    amt = float(input("Enter transfer amount: "))
    if acc["balance"] - amt < MIN_BALANCE:
        print("Minimum balance must be maintained!")
        return

    acc["balance"] -= amt
    accounts[receiver]["balance"] += amt

    acc["transactions"].append("Transferred ₹" + str(amt) + " to " + receiver)
    accounts[receiver]["transactions"].append("Received ₹" + str(amt))
    print("Transfer successful!")

# Show Transaction History
def mini_statement(acc):
    print("\nMini Statement:")
    for t in acc["transactions"]:
        print(t)

# Login
def login():
    name = input("Enter name: ")
    pin = input("Enter PIN: ")

    if name in accounts and accounts[name]["pin"] == pin:
        acc = accounts[name]
        while True:
            print("\n1.Balance 2.Deposit 3.Withdraw 4.Transfer 5.Statement 6.Logout")
            choice = input("Choose: ")

            if choice == "1":
                print("Balance: ₹", acc["balance"])
            elif choice == "2":
                deposit(acc)
            elif choice == "3":
                withdraw(acc)
            elif choice == "4":
                transfer(acc)
            elif choice == "5":
                mini_statement(acc)
            elif choice == "6":
                break
            else:
                print("Invalid choice!")
    else:
        print("Login failed!")

# Main Menu
while True:
    print("\n--- Mini Banking System ---")
    print("1.Create Account 2.Login 3.Exit")
    ch = input("Enter choice: ")

    if ch == "1":
        create_account()
    elif ch == "2":
        login()
    elif ch == "3":
        print("Program Ended")
        break
    else:
        print("Invalid choice!")

```

```python
# Personal Expense Tracker

expenses = []
MONTHLY_BUDGET = 5000

# Add expense
def add_expense():
    date = input("Enter date (DD/MM/YYYY): ")
    category = input("Enter category: ")
    amount = float(input("Enter amount: "))

    expenses.append([date, category, amount])
    print("Expense added successfully!")

# View all expenses
def view_expenses():
    if not expenses:
        print("No expenses found.")
        return

    print("\nDate\t\tCategory\tAmount")
    for e in expenses:
        print(f"{e[0]}\t{e[1]}\t\t₹{e[2]}")

# Monthly summary
def monthly_summary():
    total = 0
    for e in expenses:
        total += e[2]

    print("Total Monthly Expense: ₹", total)

    if total > MONTHLY_BUDGET:
        print("⚠ Budget Breach Alert!")

# Category-wise report
def category_report():
    category_total = {}

    for e in expenses:
        category = e[1]
        amount = e[2]

        if category in category_total:
            category_total[category] += amount
        else:
            category_total[category] = amount

    print("\nCategory-wise Spending:")
    for c in category_total:
        print(c, ":", "₹", category_total[c])

# Highest expense day
def highest_expense_day():
    highest = 0
    day = ""

    for e in expenses:
        if e[2] > highest:
            highest = e[2]
            day = e[0]

    print("Highest spending day:", day, "Amount: ₹", highest)

# Dominant spending category
def dominant_category():
    category_total = {}

    for e in expenses:
        if e[1] in category_total:
            category_total[e[1]] += e[2]
        else:
            category_total[e[1]] = e[2]

    max_cat = ""
    max_amt = 0

    for c in category_total:
        if category_total[c] > max_amt:
            max_amt = category_total[c]
            max_cat = c

    print("Dominant Spending Category:", max_cat, "₹", max_amt)

# Main menu
while True:
    print("\n--- Personal Expense Tracker ---")
    print("1. Add Expense")
    print("2. View Expenses")
    print("3. Monthly Summary")
    print("4. Category-wise Report")
    print("5. Highest Expense Day")
    print("6. Dominant Category")
    print("7. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_expense()
    elif choice == "2":
        view_expenses()
    elif choice == "3":
        monthly_summary()
    elif choice == "4":
        category_report()
    elif choice == "5":
        highest_expense_day()
    elif choice == "6":
        dominant_category()
    elif choice == "7":
        print("Program Ended")
        break
    else:
        print("Invalid choice!")
```
