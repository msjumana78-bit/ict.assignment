```python
#Write a Python function to sum all the numbers in a list.

def sum(numbers):
  total=0
  for x in numbers:
    total+=x
  return total
print(sum((8,2,3,0,7)))
```

```python
#Write a program to rename a key city to a location in the following dictionary.
sample_dict = { "name": "Kelly", "age":25, "salary": 8000, "city": "New york"}

sample_dict = { "name": "Kelly", "age":25, "salary": 8000, "city": "New york"}
print(f"Original dictionary: {sample_dict}")

# Get the value associated with the 'city' key
city_value = sample_dict.pop('city')

# Add a new key 'location' with the retrieved value
sample_dict['location'] = city_value

print(f"Dictionary after renaming 'city' to 'location': {sample_dict}")
```
